Print all three arrays. Explain what happened.

```python
arr = np.array([1,2,3,4])
print("original array :",arr)

view_arr = arr.view()
copy_arr = arr.copy()

arr[0]=10
print("after changing the original array")
print("original array :",arr)
print("View array:- ", view_arr)
print("Copy array:- ", copy_arr)
print("after changing the original array the view array was alos changes as view shares the memory refernce with the original ,\n where copy greates a new memory refernce and copy the elemts ")
```

Q11.

Given: arr = np.array([[1,2,3], [4,5,6]])
Apply:

· flatten()
· ravel()
Modify first element of both results. Observe difference.

```python
arr = np.array([[1,2,3,4],[4,5,6,8]])
print(arr)
arr1 = arr.flatten()
arr1[3]=100
print(arr, arr1)
arr = np.array([[1,2,3,4],[4,5,6,8]])
arr2 = arr.ravel()
arr2[3] = 100
print(arr, arr2)

print("\n flastten doesn't change the first arr as it duplicated the array , where the ravel just passs the refernce and the original array also got modified by changing the new array created with ravel Method")
```

Q12.

Create array from 1 to 12. Split into 3 equal parts. Delete last element from each part. Combine all remaining elements into a single array.

```python
arr = np.arange(1,13)

arr1 = np.split(arr,3)
combine_array = []

print(arr1)
print(type(arr1))

arr2 = [split[:-1] for split in arr1]
print(arr2)

for split in arr1:
    combine_array.extend(split)

print(combine_array)
combine_array = np.array(combine_array)
print(combine_array)
print(type(combine_array))
```
