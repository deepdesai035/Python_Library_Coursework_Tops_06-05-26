13. Create a 2×5 matrix using np.random.randn() and compute:

```python
np.random.seed(42)

arr = np.random.randn(10).reshape(2,5)
print(arr)

print("Total Memrory Used- " , arr.nbytes)
print("Size of each element:- ", arr.itemsize)

```

14. Create an 3x3 array of numbers from 1 to 9 and:

    · Flatten it using ravel()
    · Flatten it using flatten()
    . Explain the difference

```python
arr = np.arange(1,10).reshape(3,3)
print(arr)

print("--Ravel--")
arr_ravel = arr.ravel()
arr_ravel[0] =100
print(arr)
print(arr_ravel)

print("--flatten--")
arr = np.arange(1,10).reshape(3,3)
arr_flatten = arr.flatten()
arr_flatten[0] = 200
print(arr)
print(arr_flatten)


print("\n\n from the Output we can see that the ravel has changed the original array where as the flatten has not changed the original array")
```

15. Create an empty 3×7 matrix and Transpose.

```python
arr = np.empty((3,7))
print(arr)
arr_trans = arr.T
print(arr_trans)
```
