Question 15 - Mini Data Cleaning & Analysis Task

Create a dataset:

Customer, Age, City, Price, Quantity
Include:

. Missing values in Age and City
. Duplicate rows

Tasks:

1 Clean missing values
. Replace Age with mean
. Replace City with "Unknown"

2 Remove duplicate rows

3 Create new column:
Revenue = Price * Quantity

4 Filter rows where Revenue > 500

5 Sort the dataset by Revenue in descending order

6 Reset the index

7 Identify any outliers in Revenue using IQR method

```python
import pandas as pd 
import numpy as np 
```

```python
data = {
    "Customer": ["John","John", "Emma", "Liam", "Olivia", "Noah",
                 "Ava", "Emma", "Sophia", "James", "Mia", "Noah"],
    "Age": [25, 25, 30, np.nan, 28, 35,
            22, 30, np.nan, 40, 27, 35],
    "City": ["Ahmedabad","Ahmedabad","Mumbai", "Delhi", np.nan, "Bangalore",
             "Ahmedabad", "Mumbai", "Pune", "Chennai", np.nan, "Bangalore"],
    "Price": [500,500, 1200, 800, 1500, 2000,
              450, 1200, 950, 5000, 700, 2000],
    "Quantity": [2,2, 1, 3, 2, 5,
                 1, 1, 2, 10, 2, 5]
}

dataset = pd.DataFrame(data)

dataset
```

```python
# task:1 Cleaning the missig values :- age with mean and city with unknown
age_column = dataset["Age"]

print(age_column.isnull().sum())

city_column = dataset["City"]

print(city_column.isnull().sum())
```

```python
# dataset["Age"] = dataset["Age"].fillna(dataset["Age"].mean())
# dataset["City"] = dataset["City"].fillna(dataset["City"].mode()[0])

dataset

dataset = dataset.fillna({
    "Age" : dataset["Age"].mean(),
    "City" : dataset["City"].mode()[0]
})

dataset
```

 ## 2. Remove duplicate rows

```python
print(dataset)

dataset = dataset.drop_duplicates()

print(dataset)
```

## 3. Create new column:
Revenue = Price * Quantity

```python
dataset["Revenue"] = dataset["Price"] * dataset["Quantity"]

dataset
```

```python
dataset
```

## 4. Filter rows where Revenue > 500

```python
revenue_data = dataset[dataset["Revenue"] > 500]
print(revenue_data)
```

## 5. Sort the dataset by Revenue in descending order


```python
dataset = dataset.sort_values (by = "Revenue" , ascending= False)
dataset
```



## 6. Reset the index



```python
dataset

sorted_dataset = dataset.sort_values(by = "Revenue")
sorted_dataset.reset_index()

sorted_dataset.reset_index(drop= True)
```


## 7. Identify any outliers in Revenue using IQR method

```python
Q1 = dataset["Revenue"].quantile(0.25)
Q2 = dataset["Revenue"].quantile(0.50)
Q3 = dataset["Revenue"].quantile(0.75)

lower_bound = Q1 - 1.5*(Q3-Q1)
upper_bound = Q3 + 1.5*(Q3-Q1)

if lower_bound <0:
    lower_bound = 0

print(lower_bound , upper_bound)
```

```python
dataset[(dataset["Revenue"] < lower_bound ) | (dataset["Revenue"]>upper_bound)]
```
