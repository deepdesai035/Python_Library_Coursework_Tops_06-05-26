## handling data in pandas 

# handling the null values

```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Customer" : ["Aman", "Deep", "Raj", "Sneha"],
    "Age" : [25, np.nan ,40, np.nan],
    "city": ["Delhi" , "Mumbai", "Delhi" , "banglore"]
}

df = pd.DataFrame(data)
df
```

```python
# step -1 

df.isnull()
```

```python
## counting missing values 
# Step -2 
df.isnull().sum()
```

```python
#step -3 

# Drop the row having missing value 

df.dropna()
```
