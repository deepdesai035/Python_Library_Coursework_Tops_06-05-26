```python
arr = np.array(np.random.rand(10))
print(arr)

#1. reshape() — Change Shape Without Changing Data
new_arr = np.reshape(arr,(2,5))
print(new_arr)
print(arr)
```

```python
#resize() — Actually Modifies Original Array
arr= np.array(np.random.randn(15))
print(arr)
new_arr = np.resize(arr,(3,5))
print(new_arr)
arr.resize((3,5))
print(arr)
```

### Difference: view() vs copy()
| Feature              | view() | copy()          |
| -------------------- | ------ | --------------- |
| Shares memory        | Yes    | No              |
| Faster               | Yes    | Slightly slower |
| Original changes too | Yes    | No              |


### ravel() — Flatten Array (Usually View)

```python
arr = np.array(np.random.randint(0, 10, size=(5, 5)))
arr = arr.ravel()
print(arr)
arr = np.array(np.random.randint(0, 10, size=(5, 5)))
arr = arr.flatten()
print(arr)
```

```python

arr = np.array([[1,2],[3,4]])

print(arr.reshape(4))
print("------------------")

arr = np.array([[1,2],[3,4]])
arr1 = arr.ravel()
print(arr)

print(arr1)




print("------------------")


arr = np.array([[1,2],[3,4]])
arr1 = arr.flatten()
print(arr)
print(arr1)

print("------------------")

arr = np.array([[1,2],[3,4]])
print(arr.transpose())
```
